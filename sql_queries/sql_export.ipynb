```python
import pandas as pd
from sqlalchemy import create_engine

engine = create_engine('mysql+pymysql://root:root@localhost/wb_proj_doc')
sql_file_path = "C:/Users/wikku/global_china_initiative/workspace/wb_cn_lending/sql_export/"
```

# Project Document Date

```python
query = """
SELECT
p.project_id,
p.countryshortname,
p.boardapprovaldate,
d.docdt,
d.document_type,
d.document_id
FROM projects as p
JOIN documents as d ON p.project_id = d.project_id;
"""   

df = pd.read_sql(query, engine)
df.to_csv(f'{sql_file_path}project_document_date.csv', index=False)
```

# Project Sector %

```python
query = """
SELECT 
p.project_id,
psl.sector_name,
ps.sector_percent
FROM projects as p
JOIN project_sectors AS ps ON ps.project_id = p.project_id 
JOIN proj_sector_lookup AS psl ON psl.sector_code = ps.sector_code;
"""   

df = pd.read_sql(query, engine)
df.to_csv(f'{sql_file_path}project_sector_percent.csv', index=False)
```

# Project Major Sector

```python
query = """
SELECT 
p.project_id,
msl.major_sector_name
FROM projects as p
JOIN project_major_sectors as pms on pms.project_id = p.project_id
JOIN major_sector_lookup AS msl ON msl.major_sector_code = pms.major_sector_code;
"""   

df = pd.read_sql(query, engine)
df.to_csv(f'{sql_file_path}project_major_sector.csv', index=False)
```
